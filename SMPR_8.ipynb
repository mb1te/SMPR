**ЗАДАНИЕ 8. ЛИНИИ УРОВНЯ НОРМАЛЬНОГО РАСПРЕДЕЛЕНИЯ**

- Отобразить на графике линии уровня нормального распределения с указанием на них значения плотности распределения
- Расссмотреть все особые случаи

$N$-мерное нормальное распределение: $N(x, \mu, Σ) = \frac{1}{\sqrt{(2\pi})^n |\Sigma|} \exp{-\frac{1}{2}(x-\mu)^T\Sigma^{-1}(x-\mu)}, x \in \mathbb{R}^n$

$1.$ Пусть признаки не коррелируют, тогда ковариационная матрица будет диагональной. Линии уровня тогда будут эллипсоидами с центром $\mu$, будут параллельны осям и расстянуты соотвественно значениям в ковариационной матрице.

Нарисуем несколько примеров:

$\Sigma = 
\begin{pmatrix}
1 & 0\\
0 & 1\\
\end{pmatrix}$

$\Sigma = 
\begin{pmatrix}
2 & 0\\
0 & 1\\
\end{pmatrix}$

$\Sigma = 
\begin{pmatrix}
1 & 0\\
0 & 3\\
\end{pmatrix}$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal
    
matrixes = [
  [
   [1, 0],
   [0, 1]
  ],
  [
   [2, 0],
   [0, 1]
  ],
  [
   [1, 0],
   [0, 3]
  ]
]

for matrix in matrixes:
    x, y = np.mgrid[-3:3.01:.01, -3:3.01:.01]
    pos = np.dstack((x, y))
    rv = multivariate_normal([0, 0], matrix)
    plt.figure()
    CS = plt.contour(x, y, rv.pdf(pos))
    plt.clabel(CS, inline=1, fontsize=10)
```

2. Пусть признаки коррелируют. Тогда матрица ковариации не диагональна, а линии уровня это эллипсоиды, которые наклонены относительно осей координат. 

Рассмотрим $\Sigma = \begin{pmatrix}
5 & 1\\
1 & 3 \\
\end{pmatrix}$

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

matrixes = [
  [
   [5, 1],
   [1, 3]
  ]
]

for matrix in matrixes:
  x, y = np.mgrid[-4:4.01:.01, -4:4.01:.01]
  pos = np.dstack((x, y))
  rv = multivariate_normal([0, 0], matrix)
  plt.figure()
  CS = plt.contour(x, y, rv.pdf(pos))
  plt.clabel(CS, inline=1, fontsize=10)
```
